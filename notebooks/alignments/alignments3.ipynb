```python
### JUST GENERATIONS SLURMM SCRIPTS
import os

start_dir = 0
end_dir = 13

# Common SLURM script header for all indices
slurm_script_header = """#!/bin/bash
#SBATCH --job-name=md{index}
#SBATCH --output=output.md
#SBATCH --error=error.md
#SBATCH --gres=gpu:1
#SBATCH --time=50:00:00
"""

# Common SLURM script footer for all indices
slurm_script_footer = """
module load cuda/12.4
module load gnu13/13.2.0

python -c "import torch; print(torch.__version__); print(torch.cuda.is_available())"
python 0.py {arg}
"""

for i in range(start_dir, end_dir + 1):
    dir_name = f"md{i}"
    
    if not os.path.exists(dir_name):
        print(f"Directory {dir_name} does not exist. Creating it now.")
        os.makedirs(dir_name)
    
    # Specific section based on index
    if 0 <= i <= 3:
        slurm_script_specific = """#SBATCH --partition=gpucompute-a40
#SBATCH --mem=120000
#SBATCH --cpus-per-task=6
"""
    else:
        slurm_script_specific = """#SBATCH --partition=gpu-preempt
#SBATCH --mem=100000
#SBATCH --cpus-per-task=3
"""

    slurm_script_content = (
        slurm_script_header.format(index=i) +
        slurm_script_specific +
        slurm_script_footer.format(arg=i)
    )
    
    slurm_script_path = os.path.join(dir_name, "3.sh")
    
    with open(slurm_script_path, "w") as script_file:
        script_file.write(slurm_script_content)

    os.chmod(slurm_script_path, 0o755)

    print(f"SLURM script created for {slurm_script_path}")

print("All scripts have been generated.")
```

```python
### CLEANS ALL DIRECTORIES AND ADDS ONLY BASE FILES FROM MD DIRECTORY. ALSO CREATES SLURM SCRIPTS
import os
import shutil

start_dir = 0
end_dir = 13
base_dir = "md"
base_files = [
    "0.py", "1.py", "2.py", "4f3t_v4G2c.pdb", "5v2h.cif",
    "ions.mdp", "step4.0_minimization.mdp", "step4.1_equilibration.mdp",
    "step5_production.mdp"
]

for i in range(start_dir, end_dir + 1):
    dir_name = f"md{i}"
    if os.path.exists(dir_name):
        for filename in os.listdir(dir_name):
            file_path = os.path.join(dir_name, filename)
            if os.path.isfile(file_path) or os.path.islink(file_path):
                os.unlink(file_path)

for i in range(start_dir, end_dir + 1):
    dir_name = f"md{i}"
    if not os.path.exists(dir_name):
        os.makedirs(dir_name)
    for file_name in base_files:
        src_file = os.path.join(base_dir, file_name)
        dest_file = os.path.join(dir_name, file_name)
        if os.path.isfile(src_file):
            shutil.copy2(src_file, dest_file)

slurm_script_header = """#!/bin/bash
#SBATCH --job-name=md{index}
#SBATCH --output=output.md
#SBATCH --error=error.md
#SBATCH --gres=gpu:1
#SBATCH --time=50:00:00
"""

slurm_script_footer = """
module load cuda/12.4
module load gnu13/13.2.0
python -c "import torch; print(torch.__version__); print(torch.cuda.is_available())"
python 0.py {arg}
"""

for i in range(start_dir, end_dir + 1):
    dir_name = f"md{i}"
    if 0 <= i <= 3:
        slurm_script_specific = """#SBATCH --partition=gpucompute-a40
#SBATCH --mem=120000
#SBATCH --cpus-per-task=6
"""
    else:
        slurm_script_specific = """#SBATCH --partition=gpu-preempt
#SBATCH --mem=100000
#SBATCH --cpus-per-task=3
"""
    slurm_script_content = (
        slurm_script_header.format(index=i) +
        slurm_script_specific +
        slurm_script_footer.format(arg=i)
    )
    slurm_script_path = os.path.join(dir_name, "3.sh")
    with open(slurm_script_path, "w") as script_file:
        script_file.write(slurm_script_content)
    os.chmod(slurm_script_path, 0o755)
    print(f"SLURM script created for {slurm_script_path}")

print("All directories have been synchronized and scripts have been generated.")
```

```python
### RUN ALL SLURM SCRIPTS WITH 5 SEC DELAY ADDED
import subprocess
import os
import time

start_dir = 1
end_dir = 16
slurm_script_name = "3.sh"

for i in range(start_dir, end_dir + 1):
    directory = f"md{i}"
    
    if os.path.isdir(directory):
        slurm_script_path = os.path.join(directory, slurm_script_name)
        
        if os.path.isfile(slurm_script_path):
            try:
                os.chdir(directory)
                print(f"Changed to directory: {os.getcwd()}")
                print(f"Submitting {slurm_script_name}...")
                subprocess.run(["sbatch", slurm_script_name], check=True)
                print(f"Submitted {slurm_script_name} successfully.")
                # Wait for 5 seconds before moving to the next script
                time.sleep(10)
            except subprocess.CalledProcessError as e:
                print(f"Failed to submit {slurm_script_name}: {e}")
            except Exception as e:
                print(f"An error occurred: {e}")
            finally:
                os.chdir("..")  # Return to the parent directory
        else:
            print(f"SLURM script {slurm_script_name} not found in directory {directory}.")
    else:
        print(f"Directory {directory} does not exist.")

print("Script execution completed.")
```

```python
### CHECKS ALL CREATED .GRO FILES, LOOKING FOR MISMATCH BETWEEN GENE_ALIGNMENTS3 AND THE PRODUCED FILE
from pymol import cmd
import os
import csv

directory = "complete_md_amide"
files = sorted(os.listdir(directory))
csv_path = "../gene_alignments3.csv"

def check_matches(filename, target_rna):
    full_path = os.path.join(directory, filename)
    cmd.load(full_path, "structure")
    
    indices_resn = []
    cmd.iterate("structure", "indices_resn.append((resi, resn))", space={"indices_resn": indices_resn})

    last_index = None
    count = 0
    mod_strand = ""
    target_strand = ""

    for resi, resn in indices_resn:
        resi = int(resi)
        if resi != last_index:
            count += 1
            
            if count > 859 and count <= 859 + 4:
                mod_strand += resn
                if count == 859 + 4 and mod_strand != "AUR3R4":
                    return f"Mismatch in modified strand: strand from gro = {mod_strand}"
            
            elif count > 859 + 4 + 17:
                if resi <= 21:
                    target_strand += resn
                    if resi == 21:
                        if target_strand != target_rna:
                            return f"Mismatch: gro = {target_strand}, gene = {target_rna}"
                        break
            
            last_index = resi
    return None

with open(csv_path, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    ensg_dict = {row['ensembl_id']: row['target_rna'] for row in reader}

for file in files:
    if file.endswith(('.pdb', '.cif', '.gro')):
        ensg_id = file.split('_')[1].split('.')[0]
        target_rna = ensg_dict.get(ensg_id, None)
        
        if target_rna:
            result = check_matches(file, target_rna)
            if result:
                print(f"{ensg_id}: {result}")

print("End of checking.")
```
